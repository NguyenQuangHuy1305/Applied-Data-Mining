```python
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib as plt

df = pd.read_csv('raw_bank_data.csv', sep=',')

df.head()
```

```python
df.describe()
```

```python
df.isnull().any()
```

```python
v1 = df['age'].mean(skipna=True)
print(v1)
```

```python
values = {'age':v1}
clean_df = df.fillna(value=values)
clean_df.head()
```

```python
clean_df.isnull().any()
```

```python
nonan_clean_df=clean_df.dropna(axis=0,how='any')
nonan_clean_df.head()
```

```python
nonan_clean_df.isnull().any()
```

```python
nonan_clean_df_noday = nonan_clean_df.drop('day',1)
nonan_clean_df_noday_nomonth =  nonan_clean_df_noday.drop('month',1)
nonan_clean_df_noday_nomonth.head()
data=nonan_clean_df_noday_nomonth
data.isnull().any()
```

```python
data.head()
```

```python
import numpy as np
import sklearn
from sklearn.metrics import *
from sklearn.svm import SVC, LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import minmax_scale
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer 
from sklearn import preprocessing

df=data

for col_name in df.columns:
    if(df[col_name].dtype=='object'):
        df[col_name] = df[col_name].astype('category')
        df[col_name] = df[col_name].cat.codes

X = np.array(df.values[:,0:14])
y = np.array(df.values[:,-1])
print(len(df['job'].unique()),len(df['marital'].unique()),len(df['education'].unique()),len(df['default'].unique()),len(df['housing'].unique()),len(df['loan'].unique()),len(df['contact'].unique()),len(df['poutcome'].unique()))


#enc = OneHotEncoder(n_values = [12, 3, 4, 2, 2, 2, 3, 4], categorical_features=[1, 2, 3, 4, 6, 7, 8, 13])
#X = enc.fit_transform(X).toarray()

#enc = OneHotEncoder(categories=[range(12), range(3), range(4), range(2), range(2), range(2), range(3), range(4)], ColumnTransformer = [1, 2, 3, 4, 6, 7, 8, 13])

ct = ColumnTransformer([('one_hot_encoder', OneHotEncoder(),[1, 2, 3, 4, 6, 7, 8, 13])], remainder="passthrough")
X = ct.fit_transform(X)

print(X.shape)
print(X[0,:])
X_n= X     # original data
#X_n = minmax_scale(X)   # min max normalization
#X_n = preprocessing.scale(X)          # z-score normalization
print(X_n)

knn_params = {'n_neighbors': [1, 5, 10, 15]}
X_tr, X_te, ys_tr, ys_te = train_test_split( X_n, y, test_size = 0.3)
        
knn = KNeighborsClassifier()
clf = GridSearchCV(knn, knn_params, cv =2, n_jobs =1, verbose =3)
clf.fit(X_tr, ys_tr)
knn_pred = clf.predict(X_te)
accuracy = accuracy_score(ys_te, knn_pred)
print("KNN -> The best parameter is %.3f Accuracy: %.6f"%(clf.best_params_['n_neighbors'],accuracy)) 

```
